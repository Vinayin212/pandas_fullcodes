### MERGING , JOINING , CONCATINATION

#### merging of 2 dataframes

```python
employee = pd.DataFrame({
    'employee_id': [ 1, 2, 3, 4, 5],
    'name' : ['john', 'aman' , 'gotam', 'sumit', 'vinay'],
    'department': ['hr', 'operation', 'marketing', 'manager', 'data scentist']
})

salaries =pd.DataFrame({
    'employee_id':[ 1, 2, 3, 6, 7],
    'salary': [30000, 40000, 50000, 60000, 100000],
    'bonus': [20000, 30000, 40000, 30000, 60000]
})
```

```python
employee
```

```python
salaries
```

```python
pd.merge(employee,salaries,on= 'employee_id', how='inner')
```

```python
pd.merge(employee,salaries,on= 'employee_id', how='outer')
```

```python
pd.merge(employee,salaries,on= 'employee_id', how='left')
```

```python
pd.merge(employee,salaries,on= 'employee_id', how='right')
```

### groupby aggregation

```python
data = {
    'category': ['a', 'b', 'c', 'a', 'b', 'a', 'b'],
    'store' : [ 's1', 's2', 's2', 's1', 's1', 's2', 's1'],
    'sales': [100, 200, 300, 100, 150, 250, 100],
    'quantity': [10, 20, 15, 12, 15, 8, 17],
    'date': pd.date_range('2026-2-2', periods=7)
}
df =pd.DataFrame(data)
```

```python
df
```

```python
cat= df.groupby('category')['sales'].sum()
cat
```

```python
cat= df.groupby('store')['sales'].sum()
cat
```

```python
cat= df.groupby(['store', 'category'])['sales'].sum()
cat
```

### pivot tables

```python
data ={
    'date': pd.date_range('2026-2-2', periods=20),
    'product': (['A', 'B', 'C'] * 7) [:20],
    'region': ['north', 'south', 'east', 'west'] * 5,
    'sale': np.random.randint(100, 1000, 20),
    'unit': np.random.randint(1, 10, 20),
    'rep' : ['john', 'bob', 'mary', 'alice'] * 5,


    }

df= pd.DataFrame(data)
df['month'] = df['date'].dt.month_name()
df['quarter'] = '0' + df['date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values='sale', index='region', columns='product', aggfunc='median')
```

```python
pivot2= pd.pivot_table(df,values=['sale','unit'], index='region', columns='product', )
pivot2
```

#### cross tabs

```python
pd.crosstab(df['region'], df['product'], )
```
